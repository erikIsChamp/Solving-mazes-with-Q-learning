```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_comparison(model_list, labels):
    # Compares the training data from selected models in a plot
    avg_rewards_list = [liste[0] for liste in model_list]
    avg_steps_list = [liste[1] for liste in model_list]

    if len(avg_rewards_list) != len(avg_steps_list) or len(avg_rewards_list) != len(labels):
        raise ValueError("All input lists must have the same length.")

    plt.figure(figsize=(10, 5))

    # Plot average rewards
    plt.subplot(1, 2, 1)
    for rewards, label in zip(avg_rewards_list, labels):
        plt.plot(rewards, label=label)
    plt.xlabel('Episode')
    plt.ylabel('Cumulative Reward')
    plt.ylim(-10000, 200)
    plt.title('Average Reward per Episode')
    plt.legend()

    # Plot average steps
    plt.subplot(1, 2, 2)
    for steps, label in zip(avg_steps_list, labels):
        plt.plot(steps, label=label)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel('Episode')
    plt.ylabel('Steps Taken')
    plt.ylim(-50, 2000)
    plt.title('Average Steps per Episode')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
# Load data from .npy files
path_folder = 'training_data/'
model_1 = np.load(path_folder + 'model_1_training_data.npy')
model_2 = np.load(path_folder + 'model_2_training_data.npy')
model_3 = np.load(path_folder + 'model_3_training_data.npy')
```

```python
plot_comparison(
    [model_1, model_2, model_3],
    labels=["model 1", "model 2", "model 3"]
)
```
